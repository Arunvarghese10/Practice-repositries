```python
## Cross validation for LGBM

```

```python
from sklearn.model_selection import KFold 
import lightgbm as lgb
 
# Similarly LGBMRegressor can also be imported for a regression model.
from lightgbm import LGBMClassifier
```

```python
kfold=KFold(n_splits=5, shuffle=True)

lgbm = LGBMClassifier(n_estimators=500, 
                     max_depth=10,
                     random_state=42,
                     learning_rate = 0.01,
                     scale_pos_weight =3)

pred_df = pd.DataFrame()
n=5
for i in range(n):
    folds = next(kfold.split(x)) ## next helps you iterate over next fold
    xtrain=x.iloc[folds[0]]  # builds x train data 
    ytrain=y.iloc[folds[0]] # creates y values 
    lgbm.fit(xtrain,ytrain)
    pred_df[i]=lgbm.predict_proba(newtest)[:,1]
```

```python
median_prob=pred_df.median(axis=1)
```

```python
submission2=pd.DataFrame({'Patient_ID':test.Patient_ID,'Health_Camp_ID':test.Health_Camp_ID,
                       'Outcome':median_prob})
```

```python
submission2.to_csv('Votingmodel2.csv', index=False)
```
